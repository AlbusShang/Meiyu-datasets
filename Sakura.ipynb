##1. Drawing figures on these two files.

###1.1 separately

```python
with sns.axes_style("ticks"):
  plt.figure(figsize=(10,4))
  plt.plot(sakura_kyoto['YEAR'], sakura_kyoto['VALUE'],marker='o',alpha=0.7,linestyle='None',markersize = 3.0,color='r')
  plt.xlabel('Year')
  plt.ylabel('Blossoming Date')
  plt.title('Kyoto Cherry Blossoming Date')
  plt.savefig(DOCU_Path/'Kyoto sakura.png')
```

```python
with sns.axes_style("ticks"):
  plt.figure(figsize=(10,4))
  plt.plot(sakura_tokyo['YEAR'], sakura_tokyo['VALUE'],marker='o',alpha=0.7,linestyle='None',markersize = 3.0,color='r')
  plt.xlabel('Year')
  plt.ylabel('Blossoming Date')
  plt.title('Tokyo Cherry Blossoming Date')
  plt.savefig(DOCU_Path/'Tokyo sakura.png')
```

###1.2 Together

```python
with sns.axes_style("ticks"):
  plt.figure(figsize=(10,4))
  plt.plot(sakura_kyoto['YEAR'], sakura_kyoto['VALUE'],marker='o',alpha=0.7,linestyle='None',markersize = 3.0)
  plt.plot(sakura_tokyo['YEAR'], sakura_tokyo['VALUE'],marker='o',alpha=0.7,linestyle='None',markersize = 3.0)
  plt.xlabel('Year')
  plt.ylabel('Blossoming Date')
  plt.legend(['Kyoto','Tokyo'])
  plt.title('Tokyo Cherry Blossoming Date')
  plt.savefig(DOCU_Path/'Japan Sakura.png')
```
